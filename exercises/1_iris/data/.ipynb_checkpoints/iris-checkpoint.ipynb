## Data analysis

Write a function that prints the mean and another that prints the standard deviation of each attribute.

```python
import numpy as np
attributes = np.loadtxt('iris.attributes', delimiter=',', dtype= float)

print(attributes)

    
mean = np.mean(attributes)
print("Mean:", mean)


# Function to calculate and print the mean of each column using just one for loop
def print_column_means(data):
    # Number of attributes (columns)
    num_attributes = len(data[0])

    # Loop through each column to calculate and print the mean
    for col in range(num_attributes):
        column_sum = sum(row[col] for row in data)  # Sum values of the current column
        column_mean = column_sum / len(data)  # Calculate the mean for the column
        print(f"Mean of Attribute {col + 1}: {column_mean:.2f}")

# Assuming `data` is already loaded
# Call the function to calculate and print the mean of each column
print_column_means(data)
```
